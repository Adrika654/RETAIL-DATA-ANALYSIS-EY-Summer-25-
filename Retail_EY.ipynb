```python
# STEP 2: Import libraries
import pandas as pd
import numpy as np

# STEP 3: Load the dataset
df = pd.read_csv('retail_data.csv')

# STEP 4: Preview the dataset
print("First 5 rows:")
print(df.head())

# STEP 5: Dataset information
print("\nData Info:")
print(df.info())

# STEP 6: Descriptive statistics
print("\nDescriptive Stats:")
print(df.describe(include='all'))

# STEP 7: Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# STEP 8: Check for duplicate rows
print("\nDuplicate Rows:")
print(df.duplicated().sum())

# STEP 9: Unique values in categorical columns
print("\nUnique values in object columns:")
for col in df.select_dtypes(include='object'):
    print(f"{col}: {df[col].nunique()} unique values")

# STEP 10 (Optional): Save cleaned version
df.to_csv('cleaned_retail_data.csv', index=False)
files.download('cleaned_retail_data.csv')
```

```python
import pandas as pd
import gender_guesser.detector as gender

# Load your dim_customer table (export it as CSV from pgAdmin or Power BI first)
df = pd.read_csv('dim_customer.csv')

d = gender.Detector()

# Extract first name (assuming it's the first word in 'Name')
df['First_Name'] = df['name'].str.split().str[0]

# Predict gender
df['Predicted_Gender'] = df['First_Name'].apply(lambda x: d.get_gender(x) if pd.notnull(x) else None)

# Normalize prediction
df['Predicted_Gender'] = df['Predicted_Gender'].replace({
    'mostly_male': 'Male',
    'male': 'Male',
    'mostly_female': 'Female',
    'female': 'Female'
})

# Optional: only update where the predicted gender is different and not null
df['Cleaned_Gender'] = df.apply(
    lambda row: row['Predicted_Gender'] if pd.notnull(row['Predicted_Gender']) else row['Gender'],
    axis=1
)

# Replace the Gender column
df['Gender'] = df['Cleaned_Gender']
df.drop(columns=['Predicted_Gender', 'Cleaned_Gender', 'First_Name'], inplace=True)

# Export the cleaned table
df.to_csv('dim_customer_cleaned.csv', index=False)
```
